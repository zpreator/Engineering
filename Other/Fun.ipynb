```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
XKCD = False
eq = lambda x: (1+0.1/x)**x-1
if XKCD:
    with plt.xkcd():
        fig = plt.figure()
        x = np.arange(1, 365, 1)
        plt.plot(x, eq(x))
else:
    fig = plt.figure()
    x = np.arange(1, 365, 1)
    plt.plot(x, eq(x))
```
